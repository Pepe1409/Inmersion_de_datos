```python
inmuebles['Precio_M2'] = inmuebles['Precio_Millon'] / inmuebles['Area']
```

```python
inmuebles.sample()
```

```python
inmuebles.info()
```

```python
Datos_m2 = inmuebles.groupby('Barrio')['Precio_M2'].mean().reset_index().sort_values(by='Precio_M2',ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(15,6))
grafica3=sns.barplot(data=Datos_m2, x='Precio_M2', y='Barrio')
grafica3.set_title('Promedio de costo de M2 en barrios de Bogota', fontsize=20)
plt.xlabel('Promedio en millones de pesos',fontsize=14)
plt.ylabel('Barrio',fontsize=14)
plt.show()
```
